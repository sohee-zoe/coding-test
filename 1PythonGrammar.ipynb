````
M = 4
N = 3


# M번 반복할 때마다 길이가 N인 리스트 초기화
# a = M X N
a = [[0] * N for _ in range(M)]
print(a)
````

````
# a(1,1) = 5 할당
a[1][1] = 5
print(a)
````

````
# 잘못된 예시
b = [[0] * N] * M
print(b)
````

````
# b(1,1) = 5 할당 X
# b(:,1) = 5
b[1][1] = 5
print(b)



for i in range(M):
  a[i][1] = 5
print(a)
````

# set 집합
- 특징
  - 중복 허용 X
  - 순서 X
- 리스트 혹은 문자열을 이용해서 초기화
  - **set()** 함수 이용
- 혹은 중괄호 {} 안에 각 원소를 콤마 , 기준으로 삽입함으로써 초기화 할 수 있음
- 데이터의 조회 및 수정에 있어 O(1)의 시간에 처리 가능


````
# 집합 자료형 초기화 방법 1
data = set([1,1,2,3,4,4,5])
print(data)

data = {1,1,2,3,4,4,5}
print(data)
````

````
a = set([1,2,3,4,5])
b = set([3,4,5,6,7])

# 합집합
print(a | b)

# 교집합
print(a & b)

# 차집합
print(a - b)
````

````
data = set([1,2,3])
print(data)

# 새로운 원소 추가
data.add(4)
print(data)

# 새로운 원소 여러 개 추가
data.update([5,6])
print(data)

# 특정한 값을 갖는 원소 삭제
data.remove(3)
print(data)
````

- 리스트, 튜플은 순서가 있어 인덱싱을 통해 자료형의 값을 얻을 수 있다.
- 사전 자료형과 집합 자료형은 순서가 없어 인덱싱으로 값을 얻을 수 없다.
  - 사전의 key 혹은 집합의 원소(element)를 이용해 *O*(1)의 시간 복잡도로 조회된다.

- ex) 공백을 기준으로 구분된 데이터를 입력 받을 때 다음과 같이 사용

### continue 키워드
- 반복문에서 남은 코드의 실행을 건너뛰고 다음 반복을 진행하고자 할 때

````
result = 0

# 1~9까지 홀수의 합
for i in range(1,10):
  if i % 2 == 0:
    continue
  result += i

print(result)
````

### break 키워드
- 반복문을 즉시 탈출하고자 할 때

````
# 1~5 정수 차례댈 ㅗ출력

i = 1

while True:
  print('현재값', i)

  if i == 5:
    break
  
  i += 1
````

### global 키워드
- 함수 바깥에 선언된 변수 참조

````
a = 0

def func():
  global a
  a += 1

for i in range(10):
  func()

print(a)
````

````
# 리스트가 전역 변수일 때 global 없어도 호출 가능

array = [1, 2, 3, 4, 5]

def func():
  array.append(6)
  print(array)

func()
````

````
array = [1, 2, 3, 4, 5]

# 함수 안에 지역 변수가 전역 변수와 동일한 이름
# 지역변수가 우선 참조
def func():
  array = [3,4,5]
  array.append(6)
  print(array)

func()
print(array)
````

````
array = [1, 2, 3, 4, 5]

# 전역 변수를 변경하고자 할 때
def func():
  global array
  array = [3,4,5]
  array.append(6)
  print(array)

func()
print(array)
````

### 람다 표현식

````
def add(a, b):
  return a + b

print(add(3,7))


print((lambda a, b,: a + b)(3, 7))
````

# 실전에서 유용한 표준 라이브러리

- **내장 함수**: 기본 입출력 함수, 정렬함수 etc.
- **itertools**: 반복되는 형태의 데이터를 처리하기 위한 유용한 기능 제공
  - 특히 순열과 조합 라이브러리
- **heapq**: 힙(heap) 자료 구조 제공
  - 일반적으로 우선 순위 큐 기능 구현하기 위해 사용
- **bisect**: 이진 탐색(Binary Search) 기능 제공
- **collections**: 덱(deque), 카운터(counter) 등의 유용한 자료 구조 포함
- **math**: 필수적인 수학적 기능 제공
  - 팩토리얼, 제곱근 최대공약수(GCD), 삼각함수, 파이(pi)와 같은 상수 포함


````
### 자주 사용되는 내장함수

# sum()
res = sum([1, 2, 3, 4, 5])
print(res)

# min(), max()
min_res = min(7, 3, 5, 2)
max_res = max(7, 3, 5, 2)
print(min_res, max_res)


# eval() 수식 풀기
res = eval("(3+5)*7")
print(res)


# sorted()
a = [9,4,6,8,1]
res = sorted(a)
reverse_res = sorted(a, reverse=True)
print(res)
print(reverse_res)


# sorted() with key
array = [('홍길동', 35), ('이순신', 80), ('심사임당', 75)]
res = sorted(array, key=lambda x:x[1], reverse=True)
print(res)
````

````
# 순열 permutation

from itertools import permutations

data = ['A', 'B', 'C']

result = list(permutations(data, 3))
print(result)
````

````
# 조합 combination: 서로 다른 n개에서 순서에 상관없이 서로 다른 r개를 선택

from itertools import combinations

data = ['A', 'B', 'C']

result = list(combinations(data, 2))
print(result)
````

````
# 중복 허용 순열

from itertools import product

data = ['A', 'B', 'C']

result = list(product(data, repeat = 2))
print(result)
````

````
# 중복 허용 조합

from itertools import combinations_with_replacement

data = ['A', 'B', 'C']

result = list(permutations(data, 2))
print(result)
````

# Counter
- collections 라이브러리의 Counter: 등장 횟수를 세는 기능
- 리스트와 같은 반복 가능한(iterable) 객체가 주어졌을 때 *__내부의 원소가 몇 번씩 등장했는지__* 알려줌

````
from collections import Counter

counter = Counter(['red', 'blue', 'red', 'green', 'blue', 'blue'])

print(counter['blue']) # blue 등장 횟수
print(counter['green']) # green 등장 횟수
print(dict(counter)) # 사전 자료형으로 반환
````

# 최대 공약수 & 최대 공배수
- math 라이브러리의 gcd() 함수 이용

````
import math

# 최소 공배수(LCM) 구하는 함수
def lcm(a, b):
  return a * b // math.gcd(a, b)

a = 21
b = 14

print(math.gcd(21, 14)) # 최대 공약수(GCD)
print(lcm(21, 14)) # 최소공배수(LCM)
````
